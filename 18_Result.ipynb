```python
import pickle
import os
import torch

file_names = os.listdir('output/pickle')

tensor_num = [0] * 19
mel_list = [torch.zeros(80, 3000) for _ in range(19)]
lrp_list = [torch.zeros(80, 3000) for _ in range(19)]

for file_name in file_names:
    with open('output/pickle/'+file_name, 'rb') as file:
        loaded_data = pickle.load(file)
        if loaded_data['prediction'] == loaded_data['label']:
            idx = loaded_data['label']
            """
            if tensor_num[idx] < 5:
                tensor_num[idx] += 1
                mel_list[idx] += loaded_data['mel'][:,:,0]
                lrp_list[idx] += loaded_data['lrp']
            else:
                continue
            """
            tensor_num[idx] += 1
            mel_list[idx] += loaded_data['mel'][:,:,0]
            lrp_list[idx] += loaded_data['lrp']

for i in range(19):
    mel_list[i] = mel_list[i] / tensor_num[i]
    lrp_list[i] = lrp_list[i] / tensor_num[i]
```
